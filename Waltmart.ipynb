```python
import pandas as pd
```

```python
ventas = pd.read_csv(r'C:\Users\Usuario\Desktop\Walmart_sales.csv')
```

```python
ventas
```

```python
ventas = ventas.drop(columns=['Temperature'])
```

```python
titulos = ['tienda', 'fecha', 'ventas_Semanales', 'vacaciones', 'precio_combustible', 'cpi', 'desempleo']
```

```python
ventas.columns = titulos
```

```python
ventas.isna().sum()
```

```python
ventas['fecha'] = pd.to_datetime(ventas['fecha'], format='%d-%m-%Y')
```

```python
ventas.describe()
```

```python
ventas_globales = ventas.groupby('fecha')['ventas_Semanales'].sum().reset_index()
```

```python
ventas_globales.rename(columns={'ventas_Semanales':'ventas_globales'})
```

```python
promedio_combustible = ventas.groupby('fecha')['precio_combustible'].mean().reset_index()
```

```python
ventas_globales = pd.merge(ventas_globales, promedio_combustible, on='fecha')
```

```python
ventas_globales
```

```python
cpi_promedio = ventas.groupby('fecha')['cpi'].mean().reset_index()
```

```python
cpi_promedio
```

```python
ventas_globales = pd.merge(ventas_globales, cpi_promedio, on='fecha')
```

```python
ventas_globales
```

```python
desempleo_promedio = ventas.groupby('fecha')['desempleo'].mean().reset_index()
```

```python
desempleo_promedio
```

```python
ventas_globales = pd.merge(ventas_globales, desempleo_promedio, on='fecha')
```

```python
ventas_globales
```

```python
vacaciones = ventas.groupby('fecha')['vacaciones'].median().reset_index()
```

```python
vacaciones
```

```python
vacaciones['vacaciones'] = vacaciones['vacaciones'].astype(int)
```

```python
ventas_globales = pd.merge(ventas_globales, vacaciones, on='fecha')
```

```python
ventas_globales = ventas_globales.sort_values('fecha')
```

```python
ventas_globales.to_excel(r'C:\Users\Usuario\Desktop\walmart_final.xlsx', index=False)
```
